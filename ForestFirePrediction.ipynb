#Traintest Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape )
```

#Preprocessing

```python
from sklearn.preprocessing import Normalizer
```

```python
scaler = Normalizer().fit(x_train)
Normalizer_x = scaler.transform(x_train)
Normalizer_x_test = scaler.transform(x_test)
```

#Apply Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

#Check Accuracy Training

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_pre = model.predict(x_train)

training_data_accuracy = accuracy_score(x_train_pre,y_train)
```

```python
training_data_accuracy
```

#Check Accuracy Testing

```python
x_test_pre = model.predict(x_test)

testing_data_accuracy = accuracy_score(x_test_pre,y_test)
```

```python
testing_data_accuracy
```
